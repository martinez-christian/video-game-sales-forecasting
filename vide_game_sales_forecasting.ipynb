## Handling Missing Values

### `year_of_release`
- **Reason for missingness:** Some games may not have had an official release year recorded (especially older or poorly documented titles).  
- **Decision:** Dropped these rows, since the release year is essential for analyzing trends over time.

---

### `user_score`
- **Reason for missingness:** Many entries contained `"tbd"` (‚Äúto be determined‚Äù), meaning the game had not received enough user reviews at the time of data collection.  
- **Decision:** Dropped these rows to avoid misleading analyses, since imputing values would artificially distort user sentiment.

---

### `critic_score`
- **Reason for missingness:** Smaller or niche games were sometimes never reviewed by critics.  
- **Decision:** Dropped these rows, as including them without valid scores would bias the comparison between critic and user ratings.

---

### `rating`
- **Reason for missingness:** ESRB ratings are not always available for games released outside North America, or for older titles released before the ESRB system.  
- **Decision:** Filled missing values with `"Unknown"` to retain these rows while clearly indicating that a rating is unavailable.

---

**Summary:**  
I dropped rows where missingness made the record unusable (`year_of_release`, `user_score`, `critic_score`). For `rating`, I filled with `"Unknown"` when the absence was due to regional or historical differences, allowing the rest of the record to remain useful.

## 3.2 Platform Sales Analysis Over Time

Now let's analyze how sales vary across platforms and years:

```python
# Calculate total sales by platform and year
sales_by_platform_year = (
    df.groupby(['platform', 'year_of_release'])['total_sales']
    .sum()
    .reset_index()
    .sort_values(['year_of_release', 'platform'])
)

print(sales_by_platform_year.head(15))
```

```python
# Create a heatmap of platform sales over time

import seaborn as sns
import matplotlib.pyplot as plt

# Pivot and plot in one go
sales_pivot = df.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum', fill_value=0)

plt.figure(figsize=(14,8))
sns.heatmap(sales_pivot, cmap="YlGnBu")
plt.title("Total Sales by Platform and Year")
plt.show()
```

```python
# Identify platforms with declining sales

import seaborn as sns
import matplotlib.pyplot as plt

# Pick top 6 platforms by total sales
top_platforms = sales_by_platform_year.groupby('platform')['total_sales'].sum().nlargest(6).index

# Filter only those platforms
filtered = sales_by_platform_year[sales_by_platform_year['platform'].isin(top_platforms)]

# Plot with seaborn
plt.figure(figsize=(12,6))
sns.lineplot(data=filtered, x='year_of_release', y='total_sales', hue='platform', marker="o")

plt.title("Sales Trends for Top Platforms")
plt.xlabel("Year of Release")
plt.ylabel("Total Sales (millions)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## 3.7 Cross-Platform Comparison

Compare sales performance of games across different platforms:

```python
# Count how many unique platforms each game was released on
multi_platform = (
    df.groupby('name')['platform']
      .nunique()
      .reset_index()
      .rename(columns={'platform':'platform_count'})
)

# Filter games with more than 1 platform
multi_platform_games = multi_platform[multi_platform['platform_count'] > 1]

print(f"Games released on multiple platforms: {len(multi_platform_games)}")
multi_platform_games.head(10)
```

# Step 5 : Hypothesis Tests

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

‚ÄîHow you formulated the null and alternative hypotheses

‚ÄîWhat criteria you used to test the hypotheses~~,~~ and why



```python
from scipy import stats

# Select the two groups
xone = df_relevant[df_relevant['platform'] == 'XOne']['user_score'].dropna()
pc   = df_relevant[df_relevant['platform'] == 'PC']['user_score'].dropna()

# Welch‚Äôs t-test
t_stat, p_val = stats.ttest_ind(xone, pc, equal_var=False)

print("Xbox One vs PC Ratings")
print("t-statistic:", round(t_stat, 2))
print("p-value:", round(p_val, 4))

alpha = 0.05
if p_val < alpha:
    print("Reject H0 ‚Üí The average ratings are significantly different.")
else:
    print("Fail to reject H0 ‚Üí No evidence that average ratings differ.")
```

## Hypothesis Testing Explanation (Answering the questions)

### How the Hypotheses Were Formulated
The purpose of this test is to see whether players rate Xbox One games differently from PC games.  
- **Null Hypothesis (H‚ÇÄ):** The average user ratings for Xbox One games and PC games are the same.  
- **Alternative Hypothesis (H‚ÇÅ):** The average user ratings for Xbox One games and PC games are different.  

We used this setup because hypothesis testing always starts with the assumption that there is **no difference** (H‚ÇÄ), and then we test whether the data provides enough evidence to reject that assumption in favor of H‚ÇÅ.

### Criteria Used for the Test
To compare the two groups, we applied a **two-sample Welch‚Äôs t-test**:
- This test is appropriate because we are comparing the **means of two independent groups** (Xbox One vs PC user ratings).  
- Welch‚Äôs version was chosen since it does **not assume equal variances** between the groups, making it more reliable with real-world data where variances and sample sizes can differ.  
- The chosen significance level was **Œ± = 0.05**. This means we accept up to a 5% risk of concluding there is a difference when there really isn‚Äôt one.  

**Decision Rule:**  
- If the p-value is **less than 0.05**, we reject the null hypothesis and conclude there is a significant difference.  
- If the p-value is **greater than or equal to 0.05**, we fail to reject the null hypothesis and conclude there is no evidence of a significant difference.  


```python
# Drop missing values from both groups
action = df_relevant[df_relevant['genre'] == 'Action']['user_score'].dropna()
sports = df_relevant[df_relevant['genre'] == 'Sports']['user_score'].dropna()

# Run Welch‚Äôs t-test
t_stat, p_val = stats.ttest_ind(action, sports, equal_var=False)

print("Action vs Sports")
print("t-stat:", round(t_stat, 2), "p-value:", round(p_val, 4))

if p_val < 0.05:
    print("Reject H0 ‚Üí Action and Sports ratings differ.")
else:
    print("Fail to reject H0 ‚Üí No evidence of difference.")

print("Exact p-value:", p_val)
```

### Explanation of Hypotheses  
To compare the user ratings between *Action* and *Sports* genres, we formally state the null and alternative hypotheses as follows:

- **Null Hypothesis (H‚ÇÄ):** The average user ratings for *Action* and *Sports* games are the same.  
- **Alternative Hypothesis (H‚ÇÅ):** The average user ratings for *Action* and *Sports* games are different.  

The null hypothesis assumes no difference between the two genres, while the alternative hypothesis allows for the possibility that they are rated differently by users.  

---

### Criteria for Testing  
- **Test chosen:** Welch‚Äôs *t-test* was used because the Action and Sports samples have different sizes and potentially unequal variances. This test is more robust than the standard Student‚Äôs t-test in such cases.  
- **Significance level (Œ±):** 0.05.  
- **Decision rule:**  
  - If *p-value < 0.05*, reject the null hypothesis and conclude that Action and Sports ratings differ.  
  - If *p-value ‚â• 0.05*, fail to reject the null hypothesis, meaning there is not enough evidence to conclude a difference exists.
